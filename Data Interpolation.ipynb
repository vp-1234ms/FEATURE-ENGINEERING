## Data Interpolation
Data interpolation is the process of estimating unknown values within a dataset based on the known values. In Python, there are various libraries available that can be used for data interpolation, such as NumPy, SciPy, and Pandas. Here is an example of how to perform data interpolation using the NumPy library:

1. Linear Interpolation
2. Cubic Interpolation With Scipy
3. Polynomial Interpolation

1.Linear Interpolation

```python
import numpy as np
```

```python
x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
##Interpolate the data using linear interpolatiob
x_new=np.linspace(1,5,10) #Create new x values between 1 to 5 and 10 such values
```

```python
y_interp=np.interp(x_new,x,y)
```

```python
x_new
```

```python
y_interp
```

```python
plt.scatter(x_new,y_interp)
```

2. Cubic Interpolation With Scipy

```python
x=np.array([1,2,3,4,5])
y=np.array([1,8,27,64,125])
```

```python
plt.scatter(x,y)
```

```python
from scipy.interpolate import interp1d #Interp1d means creating interpolation in one dimension
```

```python
f=interp1d(x,y,kind='cubic')
```

```python
x_new=np.linspace(1,5,10)
```

```python
y_interp=f(x_new)
```

```python
x_new
```

```python
y_interp
```

```python
plt.scatter(x_new,y_interp)
```

3. Polynomial Interpolation

```python
x=np.array([1,2,3,4,5])
y=np.array([1,8,27,64,125])
plt.scatter(x,y)
```

```python
#Interpolate data using polynomial inbterpolate
```

```python
p=np.polyfit(x,y,2)
```

```python
x_new=np.linspace(1,5,10)
x_new
```

```python
y_interp=np.polyval(p,x_new)
y_interp
```

```python
plt.scatter(x_new,y_interp)
```

```python
p=np.polyfit(x,y,4)
```

```python
y_interp=np.polyval(p,x_new)
y_interp
```

```python
plt.scatter(x_new,y_interp)
```
